```python
#importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading And Viewing Dataset
df = pd.read_csv('pubg-weapon-stats.csv')
```

```python
df
```

```python
#Dropping 1 row (Crossbow) which has NULL Values and is a outlier
df = df[df['Weapon Name'] != 'Crossbow']
```

```python
#Filling the NULL Values in the Follwoing Columns as Melee
df[['Bullet Type', 'Range', 'Bullet Speed']] = df[['Bullet Type', 'Range', 'Bullet Speed']].fillna('Melee')
```

```python
df
```

```python
#Using The Interpolate function to fill the NULL Values of the following Columns
df[['BDMG_0', 'BDMG_1', 'BDMG_2', 'BDMG_3', 'HDMG_0', 'HDMG_1', 'HDMG_2', 'HDMG_3']] = df[['BDMG_0', 'BDMG_1', 'BDMG_2', 'BDMG_3', 'HDMG_0', 'HDMG_1', 'HDMG_2', 'HDMG_3']].interpolate(method='linear')
```

```python
df
```

```python
#No NULL Values
df.isna().sum()
```

```python
#Identifying Statistics of The Dataset
df.describe()
```

```python
#Filtering to Get Weapon Type & Bullet Type
df2 = df[['Weapon Name','Weapon Type', 'Bullet Type']]
```

```python
df2
```

```python
#Creating the COUNT for the two Columns for Visualizations
Weapon_Counts = df['Weapon Type'].value_counts()
Bullet_Counts = df['Bullet Type'].value_counts()
```

```python
#Weapon Type Comparasion using Bar Chart
plt.figure(figsize=(10, 5))
Weapon_Counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Weapon Type Distribution")
plt.xlabel("Weapon Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Bullet Type Distribution 
plt.figure(figsize=(10, 5))
Bullet_Counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Bullet Type Distribution")
plt.xlabel("Bullet Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
